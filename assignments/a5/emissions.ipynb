```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy

csv_path = os.path.join("csv")

for csv in sorted(os.listdir(csv_path)):
    data = pd.read_csv(os.path.join(csv_path, csv))

    plt.figure(figsize=(10, 6))
    plt.bar(data['task_name'], data['emissions'])
    plt.xlabel('Task Name')
    plt.ylabel('Emissions (CO₂eq)')
    plt.title(f'Approx. emissions per task for {csv}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
